```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "study_hours":[1,2,3,4,5,6,7,8,20]
})
data
```

```python
data.describe()
```

# IQR METHOD

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1
LB = Q1 - 1.5 *IQR
UB = Q3 + 1.5*IQR

outliers = data[(data<(LB)) | (data > (UB))]
Q1
```

# Z-score method

```python
data = pd.DataFrame({
    "study_hours":[1,2,3,4,5,6,7,8,20]
})
data
```

```python
z_score = np.abs(stats.zscore(data["study_hours"]))
outliers = data[z_score > 2]
outliers
```

```python
plt.scatter(range(len(data)),data["study_hours"])
plt.xlabel("index")
plt.ylabel("study_hours")
plt.title("scatter plot showing outliers")
plt.show()
```

```python
plt.hist(data["study_hours"],bins = 6)
plt.xlabel("index")
plt.ylabel("study_hours")
plt.title("histogram showing outliers effect")
plt.show()
```
