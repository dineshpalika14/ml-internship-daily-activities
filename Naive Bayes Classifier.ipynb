## Naive Bayes
- Naive bayes is a supervised machine learning algorithm, which means it learns from labeled data.
- It is mainly used for classification problems, such as spam or not spam, play vs not play, and positive vs negative.
- The algorithm works using bayes' Theroem, which helps in calculating probabilities.
- Naive Bayes calculates the probability for each class.
- The class within the hghest probability is selected as the final prediction.
- It is called "naive" because it assumes that all features are independent of each other, even though this is often not true in real life. 

## Bayes' Theorem
- Bayes Theorem is a fundamental concept that forms the basis of the Naive Bsyes algorithm, It calculates the posterior probability of an event occuring based on prior knowledge and evidence.

                                       P(A∣B)=P(B∣A)P(A) / P(B)

- P(A∣B): Posterior Probability

- P(B∣A): Likelihood probability

- P(A) : Prior Probability

- P(B) : Marginal / Evidance probability 




```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\weather_Navie_bayes_dataset.csv")
df
```

```python
le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform(df[col])
df
```

```python
x = df.drop("PlayTennis", axis = 1)
y = df["PlayTennis"]
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Predicted values:", y_pred)
y_test
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model accuracy:", accuracy)
```

```python
new_weather = pd.DataFrame(
    [[2, 0, 0, 1, 1]],
    columns= ["Day", "Outlook", "Temperature", "Humidity", "Wind"],
)
probability = model.predict_proba(new_weather) [0]
probability
```

```python
labels = ["No", "Yes"]
plt.figure()
plt.bar(labels,probability)
plt.xlabel("Play")
plt.ylabel("Probability")
plt.title("Naive Bayes Probability Graph (play Tennis)")
plt.show()
```
