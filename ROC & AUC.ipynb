## ROC & AUC
### What is ROC curve?
- ROC(Receiver Operating Characteristic) curve is a graph that shows how well a binary classification model can seperate positive and negative classes at different thresold values.

#### It Plots:
- X-ais : False Positive Rate (FPR)
- Y-ais : True Positive Rate (TPR)

#### What is AUC?
- AUC (Area Under  the ROC Curve) is a single number that summarizes the ROC curve.
- AUC = 1.0 - Perfect model
- AUC = 0.5 - Random guessing 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
y_true = np.array([1,1,0,0,1])
y_scores = np.array([0.95, 0.85, 0.70, 0.40, 0.30])
```

```python
fpr, tpr, tnresholds =roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

print("AUC:", roc_auc)
```

```python
plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0,1],[0,1], linestyle='--')
plt.xlabel("False Positive Rate") 
plt.ylabel("True Positive Rate")
plt.title('ROC curve')
plt.legend()
plt.show()
```
