```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({"age": [18,22,25,30,35,40,45]})
data
```

```python
data["age_bins"] = pd.cut(data["age"],bins =4)
data
```

```python
plt.hist(data["age"],bins = 3)
plt.xlabel("age")
plt.ylabel("age_bins")
plt.title("eqaal-width binning using pd.cut")
plt.show()
```

```python
plt.scatter(range(len(data)),data["age"])
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("scatter plot showing binning")
plt.show()
```

```python
data["age_frequency_bins"] = pd.qcut(data["age"],q=4)
data
```
