## Logistic Regression:
- Logistic Regression is a supervised machine learning algorithm used for classification, prredicting the probobility of a categorical outcome (like yes/no, spam/not spam) by fitting data to a sigmoid (S-shaped) curve, which maps linear combinations of features to probabilities between zero and one.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = pd.DataFrame({
    "hours_studied": [1,2,3,4,5],
    "result": [0,0,0,1,1]
})
data
```

```python
x = data[["hours_studied"]]
y = data["result"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion_matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
hours = [[6]]
prediction = model.predict(hours)
probability = model.predict_proba(hours)

print("Prediction:", prediction)
print("Probability:", probability)
```

```python
print("w0 (intercept):", model.intercept_)
print("w1 (coefficient):", model.coef_)
```
